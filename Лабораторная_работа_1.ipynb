```python
import matplotlib.pyplot as plt
arr = ['001.dat','002.dat','003.dat','004.dat','005.dat']

for i in range(len(arr)):
    f=open(arr[i])
    X=[]
    Y=[]
    for j in range(int(f.readline())):
        x,y = f.readline().split()
        X.append(float(x))
        Y.append(float(y))
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    plt.scatter(X,Y)
    plt.title('Number of points: ' + str(len(X)))
    plt.axis('scaled')
    plt.savefig('Эпизод 1'+ 'Number of points: ' + str(len(X)))

```

```python
import matplotlib.pyplot as plt
import numpy as np
X=[]
Y=[]
k=1
with open('frames.dat') as file:
    storage = file.readlines()

plt.figure(figsize=(20, 20))


for i in range(0,len(storage),2):
    X=[]
    Y=[]
    X=storage[i].split(' ')
    X=list(map(float,X))

    Y = storage[i+1].split(' ')
    Y = list(map(float, Y))

    plt.subplot(len(storage)//4,len(storage)//4 + len(storage)%4,k)
    plt.plot(X,Y)
    plt.title('Frame ' + str(k))
    plt.grid()
    plt.xlim(0,16)
    plt.ylim(-12,12)
    k+=1

plt.savefig('Эпизод 2')
```

**Эпизод 3**

Однажды один фаĸультет сдавал один зачёт. Вот здесь лежит файлиĸ с результатами. В ĸаждой строĸе файла - результат одного студента, три поля через точĸу с запятой. Первое поле - ĸому сдавал студент. Второе поле - номер группы студента. Третье поле - оценĸа. (Имён студентов в файле нет. Потому что ĸ задаче они не относятся.)

*Задача* - постройте графиĸи распределения оценоĸ по препам и по группам.

**Формат графиĸа** - любой достаточно наглядный (stacked bar chart, grouped bar chart, pie chart, иное - зависит от вашей фантазии).

```python
import matplotlib.pyplot as plt
stud_prep=[]
stud_group=[]
stud_grade=[]
with open('students.csv') as file:
    stud_info = file.readlines()
for i in range(len(stud_info)):


  a=stud_info[i].find(';')
  stud_prep.append(str(stud_info[i][:a]))
  stud_info[i]=stud_info[i][a+1:]

  b=stud_info[i].find(';')
  stud_group.append(str(stud_info[i][:b]))
  stud_info[i]=stud_info[i][b+1:]

  stud_grade.append(int(stud_info[i]))


# Marks per prep
k=1
prep=sorted(set(stud_prep))

plt.figure(figsize=(10, 6))
plt.suptitle("Marks per prep")
grades=sorted(set(stud_grade))
for i in range(len(prep)):
  grades_new=[]
  counter=[]
  sizes=[]
  for j in range(len(stud_grade)):
    if stud_prep[j]==prep[i]:
      counter.append(stud_grade[j])
  for p in range(len(grades)):
    if (counter.count(grades[p])!=0):
      sizes.append((counter.count(grades[p])/len(counter))*100)
      grades_new.append(grades[p])
  ax=plt.subplot(len(prep)//2,len(prep)//2 + len(prep)%2,k)
  ax=plt.pie(sizes, labels = grades_new)
  k+=1
plt.show()
plt.savefig('Эпизод 3 Marks per prep')
```

```python
# Marks per group
k=1
group=sorted(set(stud_group))

plt.figure(figsize=(10, 5))
plt.suptitle("Marks per group")
grades=sorted(set(stud_grade))
for i in range(len(group)):
  grades_new=[]
  counter=[]
  sizes=[]
  for j in range(len(stud_grade)):
    if stud_group[j]==group[i]:
      counter.append(stud_grade[j])
  for p in range(len(grades)):
    if (counter.count(grades[p])!=0):
      sizes.append((counter.count(grades[p])/len(counter))*100)
      grades_new.append(grades[p])
  ax=plt.subplot(len(prep)//2,len(prep)//2 + len(prep)%2,k)
  ax=plt.pie(sizes, labels = grades_new)
  k+=1
plt.show()
plt.savefig('Эпизод 3 Marks per group')
```
